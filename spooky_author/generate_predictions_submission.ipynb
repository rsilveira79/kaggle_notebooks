```python
def write_content_txt_file(self, content):
    self.clean_txt_file()
    with open(self.txtfile.absolute(), 'w') as f:
        f.write(content)
    return None

def clean_txt_file(self):
    f = open(self.txtfile.absolute(), 'r+')
    f.truncate()
    return None

## Data cleaning functions
def remove_label(self, text):
    txt = re.sub('__label__', '', text)
    return txt

def remove_new_line(self, text):
    txt = re.sub('\n', '', text)
    return txt
```
